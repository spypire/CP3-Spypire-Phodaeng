```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
from google.colab import files
import os
```

```python
uploaded = files.upload()  # à¹€à¸£à¸µà¸¢à¸à¹ƒà¸Šà¹‰à¸•à¸±à¸§à¸­à¸±à¸à¹‚à¸«à¸¥à¸”à¸£à¸¹à¸›à¸‚à¸­à¸‡ Colab
if not uploaded:
    raise FileNotFoundError("No file uploaded!")
test_image_path = next(iter(uploaded.keys()))  # à¹€à¸à¹‡à¸šà¸Šà¸·à¹ˆà¸­à¹„à¸Ÿà¸¥à¹Œà¹€à¸‚à¹‰à¸²à¸ªà¸¹à¹ˆà¸•à¸±à¸§à¹à¸›à¸£
```

```python
# à¹‚à¸«à¸¥à¸”à¹‚à¸¡à¹€à¸”à¸¥ best
model_path = '/content/runs/detect/train/weights/best.pt'
if not os.path.exists(model_path):
    raise FileNotFoundError(f"YOLO model file not found at {model_path}")
model = YOLO(model_path)
```

```python
!pip uninstall -y numpy opencv-python opencv-python-headless torch torchvision torchaudio ultralytics
!pip install numpy==1.26.4 torch torchvision torchaudio ultralytics opencv-python-headless==4.9.0.80
```

```python
# à¹‚à¸«à¸¥à¸”à¸ à¸²à¸à¸”à¹‰à¸§à¸¢ OpenCV
image = cv2.imread(test_image_path)
if image is None:
    raise FileNotFoundError(f"Image not found at {test_image_path}")

# à¹à¸›à¸¥à¸‡à¸£à¸¹à¸›à¸ªà¸µà¹€à¸›à¹‡à¸™à¸‚à¸²à¸§à¸”à¸³
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# à¸ªà¹ˆà¸‡à¸ à¸²à¸à¹€à¸‚à¹‰à¸²à¹‚à¸¡à¹€à¸”à¸¥
results = model(image)
```

```python
for result in results:
    boxes = result.boxes.cpu().numpy()  # à¹à¸›à¸¥à¸‡à¸•à¸³à¹à¸«à¸™à¹ˆà¸‡ Bounding Box à¹ƒà¸«à¹‰à¸­à¸¢à¸¹à¹ˆà¹ƒà¸™à¸£à¸¹à¸›à¹à¸šà¸š NumPy array
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]  # à¸”à¸¶à¸‡à¸„à¹ˆà¸²à¸à¸´à¸à¸±à¸” Bounding Box (à¸¡à¸¸à¸¡à¸‹à¹‰à¸²à¸¢à¸šà¸™à¹à¸¥à¸°à¸¡à¸¸à¸¡à¸‚à¸§à¸²à¸¥à¹ˆà¸²à¸‡)
        confidence = box.conf[0]  # à¸”à¸¶à¸‡à¸„à¹ˆà¸²à¸„à¸§à¸²à¸¡à¹€à¸Šà¸·à¹ˆà¸­à¸¡à¸±à¹ˆà¸™à¸§à¸±à¸•à¸–à¸¸
        cls = int(box.cls[0])  # à¸”à¸¶à¸‡à¸„à¹ˆà¸² class id à¸‚à¸­à¸‡à¸§à¸±à¸•à¸–à¸¸à¸—à¸µà¹ˆà¸•à¸£à¸§à¸ˆà¸ˆà¸±à¸šà¹„à¸”à¹‰
        if confidence > 0.45:  # à¸à¸£à¸­à¸‡à¹€à¸‰à¸à¸²à¸°à¸§à¸±à¸•à¸–à¸¸à¸—à¸µà¹ˆà¸¡à¸µà¹€à¸Šà¸·à¹ˆà¸­à¸¡à¸±à¹ˆà¸™à¸¡à¸²à¸à¸à¸§à¹ˆà¸² 50%
            label = f"{model.names[cls]} {confidence:.2f}"  # à¸ªà¸£à¹‰à¸²à¸‡à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡à¹à¸ªà¸”à¸‡à¸Šà¸™à¸´à¸”à¸‚à¸­à¸‡à¸§à¸±à¸•à¸–à¸¸à¹à¸¥à¸°à¸„à¹ˆà¸²à¸„à¸§à¸²à¸¡à¹€à¸Šà¸·à¹ˆà¸­à¸¡à¸±à¹ˆà¸™
            # à¸•à¸µà¸à¸£à¸­à¸š (à¸ªà¸µà¹à¸”à¸‡à¹à¸¥à¸°à¸„à¸§à¸²à¸¡à¸«à¸™à¸² 3 à¸à¸´à¸à¹€à¸‹à¸¥)
            cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 3)
            # à¹€à¸‚à¸µà¸¢à¸™à¸‚à¹‰à¸­à¸„à¸§à¸²à¸¡ label
            cv2.putText(image,label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)
```

```python
# Display the resulting image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis("off")
plt.show()
```
